```python
# Function to extract name
def convert(obj):
    l = []  # List to store name
    for i in ast.literal_eval(obj):   #ast.literal_eval() convert str into list
        l.append(i['name'])
    return l
```

```python
# Apply convert function to genres
movie_final['genres'] = movie_final['genres'].apply(convert)
```

```python
# Apply convert function to keywords
movie_final['keywords'] = movie_final['keywords'].apply(convert)
```

```python
# Function to extract 3 names
def convert3(obj):
    l =[]
    counter =0
    for i in ast.literal_eval(obj):
        if counter !=3:   # Limit 3 only
            l.append(i['name'])
            counter +=1
        else:
            break
    return l
```

```python
# Apply convert3 function
movie_final['cast'] = movie_final['cast'].apply(convert3)
```

```python
# Function to extract director's name
def extract_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
# Apply extract_director function
movie_final['crew'] = movie_final['crew'].apply(extract_director)
```

```python
# String to list 
movie_final['overview'] = movie_final['overview'].apply(lambda x: x.split())
```

```python
# Removing spaces between words
movie_final['genres'] = movie_final['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movie_final['keywords'] = movie_final['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movie_final['cast'] = movie_final['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movie_final['crew'] = movie_final['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movie_final.head(3)
```

```python
# Create new feature named 'tags'
movie_final['tags'] = movie_final['genres'] + movie_final['overview'] + movie_final['keywords'] + movie_final['cast'] + movie_final['crew']
```

```python
# Select required features only
new_df = movie_final[['id', 'title', 'tags']]
new_df.head()
```

```python
# Convert list to string
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
# Convert to lower case
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# Convert the string to vector format
count_vect = CountVectorizer(max_features=5000, stop_words='english')
# Fit the dataset to the transformer
vectors = count_vect.fit_transform(new_df['tags']).toarray()
```

```python
# Cosine distance between vectors
from sklearn.metrics.pairwise import cosine_distances
similarity = cosine_distances(vectors)
similarity[1]
```

```python
# Recommendation function
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('How Do You Know')
```

```python
recommend('John Carter')
```
